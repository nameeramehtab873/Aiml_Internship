```python
Machine Learning is a way of teaching computers to learn patterns from data instead of giving them step-by-step rules like in traditional programming.
In normal coding, we provide the data and the rules, and the computer produces the output.
But in Machine Learning, we provide the data along with the correct answers, and the computer analyzes them to discover the rules by itself. 
These learned rules are called a model, and this model can then make predictions or decisions on new data. 
In simple words, Machine Learning helps computers learn from experience just like humans do.
```
