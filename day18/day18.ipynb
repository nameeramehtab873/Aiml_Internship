```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score,recall_score
```

```python
data=pd.read_csv(r"C:\Users\Admin\Downloads\Titanic-Dataset.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.Age.fillna(data.Age.mean(),inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
data.shape
```

```python
data.info()
```

```python
encoder=LabelEncoder()
```

```python
data.Sex=encoder.fit_transform(data.Sex)
data.Embarked=encoder.fit_transform(data.Embarked)
data.Cabin=encoder.fit_transform(data.Cabin)
```

```python
data.head()
```

```python
std_scaler=StandardScaler()
data.Age=std_scaler.fit_transform(data[["Age"]])
data.Fare=std_scaler.fit_transform(data[["Fare"]])
```

```python
data.head()
```

```python
x=data.drop(columns="Survived")
y=data.Survived
```

```python
x.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
base_model=DecisionTreeClassifier(max_depth=1,random_state=42)
```

```python
base_model
```

```python
classifier=AdaBoostClassifier(random_state=42,estimator=base_model,n_estimators=100)
```

```python
classifier.fit(x_train,y_train)
```

```python
pred=classifier.predict(x_test)
```

```python
pred
```

```python
accuracy_score(pred,y_test)
```

```python
recall_score(pred,y_test)
```
