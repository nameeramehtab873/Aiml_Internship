```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Scores": [50, 55, 60, 52, 58, 100, 53, 57, 59, 54]
}
df = pd.DataFrame(data)
df
```

```python
mean = df["Scores"].mean()
std = df["Scores"].std()
print("Mean (μ):", mean)
print("Standard Deviation (σ):", std)
```

```python
df["z_score"] = (df["Scores"] - mean) / std
df
```

```python
outliers = df[np.abs(df["z_score"]) > 3]
outliers
```
