```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
np.random.seed(42)
```

```python
heights = np.random.normal(loc=170, scale=10, size=1000)
incomes = np.random.exponential(scale=50000, size=1000)
scores = 100 - np.random.exponential(scale=10, size=1000)
```

```python
df = pd.DataFrame({
    "Heights": heights,
    "Incomes": incomes,
    "Scores": scores
})
```

```python
sns.histplot(df["Heights"], kde=True)
plt.title("Normal Distribution - Heights")
plt.show()

print("Mean:", df["Heights"].mean())
print("Median:", df["Heights"].median())
```

```python
sns.histplot(df["Scores"], kde=True)
plt.title("Left-Skewed Distribution - Scores")
plt.show()

print("Mean:", df["Scores"].mean())
print("Median:", df["Scores"].median())
```

```python
incomes = np.random.exponential(scale=50000, size=1000)
df = pd.DataFrame({"Household_Income": incomes})

df.head()
```

```python
sns.histplot(df["Household_Income"], kde=True)
plt.title("Right-Skewed Distribution - Household Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_value = df["Household_Income"].mean()
median_value = df["Household_Income"].median()

print("Mean:", mean_value)
print("Median:", median_value)
```
